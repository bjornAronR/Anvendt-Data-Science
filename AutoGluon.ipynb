```python
%autoreload

import pandas as pd
import numpy as np
import catboost as cb
from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
from matplotlib import pyplot as plt

import sys
sys.path.append("../src")
```

```python
def import_data(file):
    """create a dataframe from a csv file"""
    df = pd.read_csv(file, parse_dates=True, keep_date_col=True)
    return df

df_train = import_data("input/train.csv")
```

```python
train_data = TabularDataset(df_train)
```

```python
label = 'score'
time_limit = 60*30
eval_metric = 'mean_squared_error'
presets = 'good_quality'
save_path = 'agModels-predictClass2' 
save_path3 = 'agModels-predictClass3' 
```

```python
predictor = TabularPredictor(label=label, path=save_path, eval_metric=eval_metric).fit(train_data, presets=presets, time_limit=time_limit)
```

```python
predictor = TabularPredictor.load(save_path)
```

```python
predictor.leaderboard(silent=True)
```

```python
test= import_data("input/test.csv")
testX = test.drop(columns=['score'])
testy =test['score']
```

```python
predy = predictor.predict(testX)
```

```python
eval = predictor.evaluate_predictions(y_true=testy, y_pred=predy, auxiliary_metrics=True)
```

```python
eval
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(predy, testy)
print(mse)
```

```python
from itertools import combinations
# get the names of the features in top20
feature_names = top20.index.tolist()

# create a list of tuples containing all possible combinations of features in top20, but only for combinations of length 3, 4, and 5
combinations_list = [(' '.join(combination), list(combination)) for r in range(3, 6) for combination in combinations(feature_names, r)]
```

```python
top20Combo = predictor.feature_importance(data=test, features=combinations_list)
```

```python
top20Combo = top20Combo.sort_values(by='importance', ascending=False)
top_bottom_20_combo = pd.concat([top20Combo.head(20), top20Combo.tail(20)])
```
